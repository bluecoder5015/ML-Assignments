```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mu,sigma=5,1
x = np.random.normal(mu,sigma,10)
np.mean(x)
```

```python
y = 1/(np.sqrt(2*np.pi*sigma**2)) * np.exp(-0.5*((x-mu)/sigma)**2)
y
```

```python
a = sns.scatterplot(x,y,color='green')
a = sns.lineplot(x,y,color='green')
#a = sns.distplot(x,hist = False)
a.set_title("Distrution")
a.lines[0].set_linestyle("--")
a.set_xlim(0,10)

        
plt.savefig("guassian.png")
```

```python
fig,axes = plt.subplots(10,4,figsize=(13,10))
z=0
for i in range(10):
    for j in range(4):
        tmp=1/(np.sqrt(2*np.pi)) * np.exp(-0.5*((x-z))**2)
        sns.scatterplot(x,y,ax=axes[i,j])
        axes[i,j].set_xlabel(" = "+str(z))
        axes[i,j].set_xlim(0,10)
        sns.scatterplot(x,tmp,ax=axes[i,j],color="red")
        sns.lineplot(x,tmp,ax=axes[i,j],color="red")
        z=z+0.1
        
plt.savefig("plot.png")
```
